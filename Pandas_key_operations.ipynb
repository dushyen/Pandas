```python
g=df.groupby('city') ###Creates a group and store them 
g   
```

```python
h=df.groupby('temperature') ###Creates a group and store them 
h
```

```python
g.corr()
```

```python
for city, city_df in g:  ### Creates separate data frame, for each city a data frame is created, makes sure indices are inact
    print(city)  ## It in order by
    print(city_df)   ## Data frame
```

```python
for  temperature,temperature_df in h:  ### Creates separate data frame, for each city a data frame is created, makes sure indices are inact
    print(temperature)  ## It in order by
    print(temperature_df)   ## Data frame
```
