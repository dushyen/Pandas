```python
def get_many_locs(df, slices):
    arr=np.empty(0,dtype="int")
    for s in slices:
        locs=df.index.get_locs((s))
        arr=np.concatenate((arr,locs))
        
    return arr
```

```python
import pandas as pd
import json

# Multiple records with nested JSON data
data = '''
[
    {
        "name": "John",
        "age": 30,
        "address": {
            "street": "123 Main St",
            "city": "New York",
            "state": "NY"
        }
    },
    {
        "name": "Jane",
        "age": 25,
        "address": {
            "street": "456 Elm St",
            "city": "Los Angeles",
            "state": "CA"
        }
    }
]
'''

# Parse the JSON data
parsed_data = json.loads(data)

# Initialize an empty list to store the flattened data dictionaries
flat_data_list = []

# Loop through the records and create a flattened data dictionary for each record
for record in parsed_data:
    name = record['name']
    age = record['age']
    street = record['address']['street']
    city = record['address']['city']
    state = record['address']['state']
    flat_data = {
        'name': name,
        'age': age,
        'street': street,
        'city': city,
        'state': state
    }
    flat_data_list.append(flat_data)

# Create a Pandas DataFrame with the flattened data list
df = pd.DataFrame(flat_data_list)

# Print the DataFrame
print(df)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Create sample data
data = {'color': ['red', 'green', 'blue', 'green', 'red']}
df = pd.DataFrame(data)

# Convert categories to binary code
le = LabelEncoder()
df['color_bin'] = le.fit_transform(df['color'])
df['color_bin'] = df['color_bin'].apply(lambda x: np.binary_repr(x).zfill(3))

# Split binary code into separate features
df = pd.concat([df.drop('color_bin', axis=1), df['color_bin'].str.split('', expand=True).iloc[:, 1:]], axis=1)

print(df)
```

```python
import pandas as pd

# example dataframe
df = pd.DataFrame({'date': ['2022-01-01', '2022-02-01', '2022-03-01', '2023-01-01', '2023-02-01', '2023-03-01'],
                   'value': [1, 2, 3, 4, 5, 6]})

# convert date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# create new columns for year and month
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# loop over unique year and month combinations and create separate dataframes
for year in df['year'].unique():
    for month in df[df['year'] == year]['month'].unique():
        filename = f"{year}_{month}.csv"
        folder = "data"
        path = f"{folder}/{filename}"
        df_month = df[(df['year'] == year) & (df['month'] == month)]
        df_month.to_csv(path)
```

```python
import pandas as pd

# example dataframe
df = pd.DataFrame({'date': ['2022-01-01', '2022-02-01', '2022-03-01', '2023-01-01', '2023-02-01', '2023-03-01'],
                   'value': [1, 2, 3, 4, 5, 6]})

# convert date column to datetime format and set it as the index
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# loop over unique year and month combinations and create separate dataframes
for year in df.resample('Y').groups.keys():
    for month in df.resample('M').groups.keys():
        filename = f"{year.year}_{month.month}.csv"
        folder = "data"
        path = f"{folder}/{filename}"
        df_month = df.loc[(df.index.year == year.year) & (df.index.month == month.month)]
        df_month.to_csv(path)
```
